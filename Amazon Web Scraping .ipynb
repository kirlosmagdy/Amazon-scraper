```python
# get data in details for each phone (Name , main Image , other Images , color , promo , price , nunber of sells , Availability , ratings , description , reviews from egypt , reviews outside egypt)

def get_phone_data(link):
    if link == "https://www.amazon.egjavascript:void(0)":
        return {
            "category" :"unknown-category",
            "Name": "unknown-product",
            "Main Image": "No cover Image",
            "other Images": ["No other Images for the cover"],
            "colors": ["No cover colors"],
            "promo": "No Promo",
            "price before promo": "No Promo",
            "price after promo": "No Price",
            "Number of sells" : "No Data",
            "Availability" : "No Data",
            "Ratings": "No Ratings",
            "Number of Ratings": "No Ratings",
            "Description": ["No cover description"],
            "About the product": ["No About the cover data"],
            "Review from Egypt": [],
            "Review outside Egypt": []
        }
    else:
        time.sleep(random.randint(1, 5))  
        page = requests.get(link, headers=headers)
        soup = BeautifulSoup(page.content, "html.parser")

        name = soup.find(id="productTitle")

        if name is not None:
            phone_name = name.text.strip()
        else:
            phone_name = "unknown-product"

        main_phone_image = soup.find(id="imgTagWrapperId")

        if main_phone_image is not None:
            main_phone_image_link = main_phone_image.find("img").get("src")

        else:
            main_phone_image_link = "No phone Image"

        other_phone_image = soup.find(id="altImages")

        if other_phone_image is not None:
            temp_image_links = []

            for image in other_phone_image.find_all("img")[1:-1]:
                if image is not None:
                    link = image.get("src")
                    temp_image_links.append(link)

            if not temp_image_links:
                temp_image_links.append("No other Images for the phone")

            other_phone_images_link = temp_image_links

        else:
            other_phone_images_link = ["No other Images for the phone"]

        all_phone_color = soup.find(id="variation_color_name")

        if all_phone_color is not None:
            temp_phone_colors = []

            for color in all_phone_color.find_all("li"):
                img_tag = color.find("img")
                if img_tag is not None:
                    color_name = img_tag.get("alt")
                    temp_phone_colors.append(color_name)

            if not temp_phone_colors:
                temp_phone_colors.append("No phone colors")

            all_phone_color = temp_phone_colors

        else:
            all_phone_color = ["No phone colors"]

        promo = soup.find(
            "span",
            class_="a-size-large a-color-price savingPriceOverride aok-align-center reinventPriceSavingsPercentageMargin savingsPercentage",
        )

        if promo is not None:
            phone_promo = promo.get_text().strip()
        else:
            phone_promo = "No Promo" 

        price_before_promo = soup.find("span", class_="a-price a-text-price")

        if price_before_promo is not None:
            price_before_promo_text = price_before_promo.find(
                "span", class_="a-offscreen"
            )
            phone_price_before_promo = price_before_promo_text.get_text().strip()
        else:
            phone_price_before_promo = "No Promo"

        price_after_promo = soup.find(
            "span",
            class_="a-price aok-align-center reinventPricePriceToPayMargin priceToPay",
        )

        if price_after_promo is not None:
            price_after_promo_text = price_after_promo.find(
                "span", {"aria-hidden": "true"}
            )
            phone_price_after_promo = price_after_promo_text.get_text().strip()
        else:
            phone_price_after_promo = "No Price"


        no_of_phone_sells = soup.find(id="social-proofing-faceout-title-tk_bought")

        if no_of_phone_sells is not None :
            no_of_sells = no_of_phone_sells.text.strip()
        else : 
            no_of_sells = "No Data"


        phone_availability = soup.find(id="availability")

        if phone_availability is not None :
            phone_availability_text = phone_availability.text.strip()
        else :
            phone_availability_text = "No Data"
            

        ratings = soup.find("span", class_="a-icon-alt")

        if ratings is not None:
            phone_ratings = ratings.get_text().strip()
        else:
            phone_ratings = "No Ratings"

        no_of_ratings = soup.find(id="acrCustomerReviewText")

        if no_of_ratings is not None:
            phone_no_of_ratings = no_of_ratings.text.strip()
        else:
            phone_no_of_ratings = "No Ratings"

        phone_description = soup.find(
            "div", class_="a-section a-spacing-small a-spacing-top-small"
        )

        if phone_description is not None:
            temp_descriptions = []

            for data in phone_description.find_all("tr"):
                if data is not None:
                    description_text = data.text.strip()
                    temp_descriptions.append(description_text)

            if not temp_descriptions:
                temp_descriptions.append("No phone description")

            phone_description = temp_descriptions
        else:
            phone_description = ["No phone description"]

        about_the_phone = soup.find(id="feature-bullets")

        if about_the_phone is not None:

            temp_about_the_phone = []

            for data in about_the_phone.find_all("li"):
                if data is not None:
                    about_text = data.text.strip()
                    temp_about_the_phone.append(about_text)

            if not temp_about_the_phone:
                temp_about_the_phone.append("No About the phone data")

            about_the_phone = temp_about_the_phone

        else:
            about_the_phone = ["No About the phone data"]

       
        reviewer_name_and_review_content_from_egypt = soup.find_all(
            "div", class_="a-section review aok-relative"
        )

        reviews_from_egypt = []
        reviews_outside_egypt = []

        if reviewer_name_and_review_content_from_egypt:
            for content in reviewer_name_and_review_content_from_egypt:
                name_text = content.find("span", class_="a-profile-name")
                review_rating = content.find('span', class_="a-icon-alt")
                review_date = content.find('span', class_="a-size-base a-color-secondary review-date")
                review_text_content = content.find("span", class_="cr-original-review-content")

                review_info = {
                    "name": name_text.text.strip() if name_text else "No Reviewer name",
                    "rating": review_rating.text.strip() if review_rating else "No Review Rating",
                    "date": review_date.text.strip() if review_date else "No Review Date",
                    "content": review_text_content.get_text().strip() if review_text_content else "No Reviews"
                }

                review_string = f"Name: {review_info['name']}, Rating: {review_info['rating']}, Date: {review_info['date']}, Review: {review_info['content']}"
                reviews_from_egypt.append(review_string)

        reviewer_name_and_review_content_outside_egypt = soup.find_all(
            "div", class_="a-section review aok-relative cr-desktop-review-page-0"
        )

        if reviewer_name_and_review_content_outside_egypt:
            for content in reviewer_name_and_review_content_outside_egypt:
                name_text = content.find("span", class_="a-profile-name")
                review_rating = content.find('span', class_="a-icon-alt")
                review_date = content.find('span', class_="a-size-base a-color-secondary review-date")
                review_text_content = content.find("div", class_="a-expander-content reviewText review-text-content a-expander-partial-collapse-content")

                review_info = {
                    "name": name_text.text.strip() if name_text else "No Reviewer name",
                    "rating": review_rating.text.strip() if review_rating else "No Review Rating",
                    "date": review_date.text.strip() if review_date else "No Review Date",
                    "content": review_text_content.get_text().strip() if review_text_content else "No Reviews"
                }

                review_string = f"Name: {review_info['name']}, Rating: {review_info['rating']}, Date: {review_info['date']}, Review: {review_info['content']}"
                reviews_outside_egypt.append(review_string)



      
    return {
        "category" : category_name_list[0],
        "Name": phone_name,
        "Main Image": main_phone_image_link,
        "other Images": other_phone_images_link,
        "colors": all_phone_color,
        "promo": phone_promo,
        "price before promo": phone_price_before_promo,
        "price after promo": phone_price_after_promo,
        "Number of sells" : no_of_sells,
        "Availability" : phone_availability_text,
        "Ratings": phone_ratings,
        "Number of Ratings": phone_no_of_ratings,
        "Description": phone_description,
        "About the product": about_the_phone,
        "Review from Egypt": reviews_from_egypt,
        "Review outside Egypt": reviews_outside_egypt
    }
    
```

```python
# get the data of each phone and save it in DataFrame using multiprocess

with ThreadPool(processes= 6) as pool:
    get_phone_data_pipeline = pool.map(get_phone_data, all_phones_link_list)

df1 = pd.DataFrame(get_phone_data_pipeline)
```

```python
# get data in details for each cover (Name , main Image , other Images , color , promo , price , nunber of sells , Availability , ratings , description , reviews from egypt , reviews outside egypt)

def get_cover_data(link):
    if link == "https://www.amazon.egjavascript:void(0)":
        return {
            "category" :"unknown-category",
            "Name": "unknown-product",
            "Main Image": "No cover Image",
            "other Images": ["No other Images for the cover"],
            "colors": ["No cover colors"],
            "promo": "No Promo",
            "price before promo": "No Promo",
            "price after promo": "No Price",
            "Number of sells" : "No Data",
            "Availability" : "No Data",
            "Ratings": "No Ratings",
            "Number of Ratings": "No Ratings",
            "Description": ["No cover description"],
            "About the product": ["No About the cover data"],
            "Review from Egypt": [],
            "Review outside Egypt": []
        }
    else:
        time.sleep(random.randint(1, 5))  
        page = requests.get(link, headers=headers)
        soup = BeautifulSoup(page.content, "html.parser")

        name = soup.find(id="productTitle")

        if name is not None:
            cover_name = name.text.strip()
        else:
            cover_name = "unknown-product"

        main_cover_image = soup.find(id="imgTagWrapperId")

        if main_cover_image is not None:
            main_cover_image_link = main_cover_image.find("img").get("src")

        else:
            main_cover_image_link = "No cover Image"

        other_cover_image = soup.find(id="altImages")

        if other_cover_image is not None:
            temp_image_links = []

            for image in other_cover_image.find_all("img")[1:-1]:
                if image is not None:
                    link = image.get("src")
                    temp_image_links.append(link)

            if not temp_image_links:
                temp_image_links.append("No other Images for the cover")

            other_cover_images_link = temp_image_links

        else:
            other_cover_images_link = ["No other Images for the cover"]

        all_cover_color = soup.find(id="variation_color_name")

        if all_cover_color is not None:
            temp_cover_colors = []

            for color in all_cover_color.find_all("li"):
                img_tag = color.find("img")
                if img_tag is not None:
                    color_name = img_tag.get("alt")
                    temp_cover_colors.append(color_name)

            if not temp_cover_colors:
                temp_cover_colors.append("No cover colors")

            all_cover_color = temp_cover_colors

        else:
            all_cover_color = ["No cover colors"]

        promo = soup.find(
            "span",
            class_="a-size-large a-color-price savingPriceOverride aok-align-center reinventPriceSavingsPercentageMargin savingsPercentage",
        )

        if promo is not None:
            cover_promo = promo.get_text().strip()
        else:
            cover_promo = "No Promo" 

        price_before_promo = soup.find("span", class_="a-price a-text-price")

        if price_before_promo is not None:
            price_before_promo_text = price_before_promo.find(
                "span", class_="a-offscreen"
            )
            cover_price_before_promo = price_before_promo_text.get_text().strip()
        else:
            cover_price_before_promo = "No Promo"

        price_after_promo = soup.find(
            "span",
            class_="a-price aok-align-center reinventPricePriceToPayMargin priceToPay",
        )

        if price_after_promo is not None:
            price_after_promo_text = price_after_promo.find(
                "span", {"aria-hidden": "true"}
            )
            cover_price_after_promo = price_after_promo_text.get_text().strip()
        else:
            cover_price_after_promo = "No Price"


        no_of_cover_sells = soup.find(id="social-proofing-faceout-title-tk_bought")

        if no_of_cover_sells is not None :
            no_of_sells = no_of_cover_sells.text.strip()
        else : 
            no_of_sells = "No Data"


        cover_availability = soup.find(id="availability")

        if cover_availability is not None :
            cover_availability_text = cover_availability.text.strip()
        else :
            cover_availability_text = "No Data"
            

        ratings = soup.find("span", class_="a-icon-alt")

        if ratings is not None:
            cover_ratings = ratings.get_text().strip()
        else:
            cover_ratings = "No Ratings"

        no_of_ratings = soup.find(id="acrCustomerReviewText")

        if no_of_ratings is not None:
            cover_no_of_ratings = no_of_ratings.text.strip()
        else:
            cover_no_of_ratings = "No Ratings"

        cover_description = soup.find(
            "div", class_="a-section a-spacing-small a-spacing-top-small"
        )

        if cover_description is not None:
            temp_descriptions = []

            for data in cover_description.find_all("tr"):
                if data is not None:
                    description_text = data.text.strip()
                    temp_descriptions.append(description_text)

            if not temp_descriptions:
                temp_descriptions.append("No cover description")

            cover_description = temp_descriptions
        else:
            cover_description = ["No cover description"]

        about_the_cover = soup.find(id="feature-bullets")

        if about_the_cover is not None:

            temp_about_the_cover = []

            for data in about_the_cover.find_all("li"):
                if data is not None:
                    about_text = data.text.strip()
                    temp_about_the_cover.append(about_text)

            if not temp_about_the_cover:
                temp_about_the_cover.append("No About the cover data")

            about_the_cover = temp_about_the_cover

        else:
            about_the_cover = ["No About the cover data"]

       
        reviewer_name_and_review_content_from_egypt = soup.find_all(
            "div", class_="a-section review aok-relative"
        )

        reviews_from_egypt = []
        reviews_outside_egypt = []

        if reviewer_name_and_review_content_from_egypt:
            for content in reviewer_name_and_review_content_from_egypt:
                name_text = content.find("span", class_="a-profile-name")
                review_rating = content.find('span', class_="a-icon-alt")
                review_date = content.find('span', class_="a-size-base a-color-secondary review-date")
                review_text_content = content.find("span", class_="cr-original-review-content")

                review_info = {
                    "name": name_text.text.strip() if name_text else "No Reviewer name",
                    "rating": review_rating.text.strip() if review_rating else "No Review Rating",
                    "date": review_date.text.strip() if review_date else "No Review Date",
                    "content": review_text_content.get_text().strip() if review_text_content else "No Reviews"
                }

                review_string = f"Name: {review_info['name']}, Rating: {review_info['rating']}, Date: {review_info['date']}, Review: {review_info['content']}"
                reviews_from_egypt.append(review_string)

        reviewer_name_and_review_content_outside_egypt = soup.find_all(
            "div", class_="a-section review aok-relative cr-desktop-review-page-0"
        )

        if reviewer_name_and_review_content_outside_egypt:
            for content in reviewer_name_and_review_content_outside_egypt:
                name_text = content.find("span", class_="a-profile-name")
                review_rating = content.find('span', class_="a-icon-alt")
                review_date = content.find('span', class_="a-size-base a-color-secondary review-date")
                review_text_content = content.find("div", class_="a-expander-content reviewText review-text-content a-expander-partial-collapse-content")

                review_info = {
                    "name": name_text.text.strip() if name_text else "No Reviewer name",
                    "rating": review_rating.text.strip() if review_rating else "No Review Rating",
                    "date": review_date.text.strip() if review_date else "No Review Date",
                    "content": review_text_content.get_text().strip() if review_text_content else "No Reviews"
                }

                review_string = f"Name: {review_info['name']}, Rating: {review_info['rating']}, Date: {review_info['date']}, Review: {review_info['content']}"
                reviews_outside_egypt.append(review_string)



      
    return {
        "category" : category_name_list[2],
        "Name": cover_name,
        "Main Image": main_cover_image_link,
        "other Images": other_cover_images_link,
        "colors": all_cover_color,
        "promo": cover_promo,
        "price before promo": cover_price_before_promo,
        "price after promo": cover_price_after_promo,
        "Number of sells" : no_of_sells,
        "Availability" : cover_availability_text,
        "Ratings": cover_ratings,
        "Number of Ratings": cover_no_of_ratings,
        "Description": cover_description,
        "About the product": about_the_cover,
        "Review from Egypt": reviews_from_egypt,
        "Review outside Egypt": reviews_outside_egypt
    }
```
